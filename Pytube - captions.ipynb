```python
# a.en: English (auto-generated), en: English (US), en-GB: English (UK)
try:
    caption = yt.captions['a.en'] or yt.captions['en'] or yt.captions['en-GB']
except:
    print('No subtitles in English')
```

```python
class Subtitle():
    def __init__(self, caption):
        self.caption = caption
        self.sublist = self.format_subtitle()
        self.text = self.compile_sub()
    
#     function to convert srt into a list of itmes in the format {"time": {time}, "subtitle": {subtitle}}
    def format_subtitle(self):
        subtitle = []
        caption = self.caption.split('\n')
        for index in range(len(caption) // 4):
            base = index * 4
#             create new item in subtitle
            subtitle.append({
                "time": caption[base + 1],
                "subtitle": caption[base + 2]
            })
        return subtitle
    
    def compile_sub(self):
        text = " ".join([section["subtitle"] for section in self.sublist])
        return text
```
